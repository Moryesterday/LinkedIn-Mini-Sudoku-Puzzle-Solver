```python
def pickMatchingDigit(x):
    return np.argmax(x) + 1
# Funzioni di attivazione
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)  # x è già l'attivazione

# Rete neurale
class NeuralNetwork:
    
    def __init__(self, input_size, hidden_size, output_size):
        # Pesi e bias
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def forward(self, X):
        # Forward pass
        self.z1 = X @ self.W1 + self.b1
        self.a1 = sigmoid(self.z1)
        self.z2 = self.a1 @ self.W2 + self.b2
        self.a2 = sigmoid(self.z2)
        return self.a2

    def backward(self, X, Y, lr=0.1):
        # Errore output
        dZ2 = (self.a2 - Y) * sigmoid_derivative(self.a2)
        dW2 = self.a1.T @ dZ2
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        # Errore layer nascosto
        dZ1 = (dZ2 @ self.W2.T) * sigmoid_derivative(self.a1)
        dW1 = X.T @ dZ1
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        # Aggiorna pesi e bias
        self.W2 -= lr * dW2
        self.b2 -= lr * db2
        self.W1 -= lr * dW1
        self.b1 -= lr * db1

    def train(self, X, Y, epochs=10000, lr=0.1):
        for i in range(epochs):
            self.forward(X)
            self.backward(X, Y, lr)
            if i % 100 == 0:
                loss = np.mean((self.a2 - Y)**2)
                print(f"Epoch {i}, Loss: {loss:.6f}")

    # Metodo predict per ottenere 0 o 1
    def predict(self, X):
        a = self.forward(X)
        return np.argmax(a, axis = 1) + 1
    
    def getLoss(self, X, Y, lr):
        self.forward(X)
        self.backward(X, Y, lr)
        loss = np.mean((self.a2 - Y)**2)
        return loss
```
